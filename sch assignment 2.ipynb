```python
import numpy as np
def f(x):
    return (np.sin(x))/x
t=10000
x=np.linspace(-t,t,20*t+1)
left=right=middle=0
for i in range(20*t):
    if x[i]==0 or x[i+1]==0:
        continue
    left+=f(x[i])
    if i<20*t:
        right+=f(x[i+1])
        middle+=f((x[i]+x[i+1])/2)
left= left/10
right=right/10
middle=middle/10
print(left," ",middle," ",right)
```

```python
def f(x):
    return np.exp(-x**2)
t=1000
x=np.linspace(-t,t,20*t+1)
I=0
for i in range(20*t):
    I=I+(f(x[i])+f(x[i+1]))/2
I=I/10
print(I)
```
